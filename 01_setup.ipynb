```python
# 2) Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

import os
OUT_DIR = '/content/drive/MyDrive/ocr/dataset_v1'
os.makedirs(OUT_DIR, exist_ok=True)
print('OUT_DIR =', OUT_DIR)
```

```python
# 3) Upload skryptów z dysku lokalnego LUB pobranie z repo.
# Tu używamy uploadera (wybierz 4 pliki 02..05).
from google.colab import files
import shutil, os

WORK = '/content/ocr_stage1'
os.makedirs(WORK, exist_ok=True)

print('Wybierz pliki: 02_charset.py, 03_text_sampler.py, 04_generate.py, 05_validate.py')
up = files.upload()
for name, data in up.items():
    with open(os.path.join(WORK, name), 'wb') as f:
        f.write(data)
print(sorted(os.listdir(WORK)))
```
